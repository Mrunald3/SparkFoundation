Using Decision Tree Classifier to predict classes of new data.

**Importing the libraries**

```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
%matplotlib inline
```

**Reading and exploring the data from file**

```python
iris=pd.read_csv("Iris.csv")
iris.head()
```

```python
iris['Species'].value_counts()
```

```python
xdf = iris.loc[:, 'SepalLengthCm':'PetalWidthCm']
```

```python
print(xdf.head())
```

```python
ydf=iris.Species
print(ydf.head())
```

**Applying Decision Tree Algorithm**

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
```

```python
dtree.fit(xdf,ydf)
```

**Visualizing the Tree**

```python
from sklearn import tree
```

```python
plt.figure(figsize=(16,8))
tree.plot_tree(dtree);
```

**Predicting for new data**

```python
x_test=[(4.2,2.9,1.6,0.2),(5.5,4.2,2.9,0.7)]
pred_dtree = dtree.predict(x_test)
print(pred_dtree)
```

Thus, we can see that the algoritm predicts correctly for any new data given.
